```python
def hamming_distance(str1, str2):
    if len(str1) != len(str2):
        raise ValueError("Strings must be of equal length")
    
    distance = 0
    for char1, char2 in zip(str1, str2):
        if char1 != char2:
            distance += 1
    return distance

# Read strings from text file
with open("dataset.txt", "r") as file:
    strings = file.read().splitlines()

# Iterate through strings and calculate Hamming distance
for i in range(len(strings)):
    for j in range(i+1, len(strings)):
        string_a = strings[i]
        string_b = strings[j]
        
        # Pad the shorter string with spaces
        max_length = max(len(string_a), len(string_b))
        string_a = string_a.ljust(max_length)
        string_b = string_b.ljust(max_length)
        
        # Calculate the Hamming distance
        distance = hamming_distance(string_a, string_b)
        
        print(f"Hamming distance between '{string_a}' and '{string_b}': {distance}")
```
